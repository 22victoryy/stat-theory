```r
sample_4m_normal=function(){
  
  s=rnorm(30, mean=10,sd=2)        # drawing 30 random numbers from N(10,2^2)
  x_bar=mean(s)
  
  return(x_bar)                  # returning the mean of my sample of size 30
}

X_bar=replicate(1000000, sample_4m_normal())

plot(density(X_bar)) 
```

f) While it is hard to comment on the exact value of n, we know that according to the CLT, the distribution of sample means converges to a normal distribution as the sample size becomes larger in the assumption that all samples are identical in size, regardless of the population disribution shape. 
Considering as n = 5 as a small sample size and n = 30 as a large sample size, we can conclude that sample sizes that are 30 or larger are sufficient for the CLT to hold. 

g) We see that the unform distribution does not have a notable skew. Most data are centered in a uniform fashion. <br>
   We see that the chi-squared distrbution of n=2 and df=2 is left-skewed. (Moderate)<br>
   We see that chi-squared distribution of n=2 and df=50 is left skewed but more centered than the second graph.    
   (Severe)<br>

   Thus, all three graphs have a different degree of skewness.<br>

   We know that the following is true for the CLT:
    * No skew, applying the CLT, will result in no skew.
    * Right skew, applying the CLT, will result in no skew.
    * Left skew, applying the CLT, will result in no skew.

   However, strongly skewed distributins like graph 2 will require larger sample sizes for the CLT to converge to a      Normal Distribution. The more the population distribution differs from being normal, the larger the sample size 
   must be.
