```r
sample_4m_normal=function(){
  
  s=rnorm(30, mean=10,sd=2)        # drawing 30 random numbers from N(10,2^2)
  x_bar=mean(s)
  
  return(x_bar)                  # returning the mean of my sample of size 30
}

X_bar=replicate(1000000, sample_4m_normal())

plot(density(X_bar)) 
```
